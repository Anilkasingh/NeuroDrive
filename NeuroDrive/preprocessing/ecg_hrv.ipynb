```python
import neurokit2 as nk
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("ecg_data.csv")
data.head()
```

```python
data['id'] = range(1,2540)
```

```python
sns.lineplot(data, x='id', y='ecg_value')
```

```python
peaks, info = nk.ecg_peaks(data['ecg_value'], 100)
```

```python
data['is_peak'] = peaks['ECG_R_Peaks']
sns.scatterplot(data, x='id', y='ecg_value', hue='is_peak')
```

```python
hrv_time = nk.hrv_frequency(peaks, 100)
hrv_time
```

```python
n, _ = nk.ecg_process(data['ecg_value'], 100)
n.head()
```

```python
sns.lineplot(n, x='ECG_Raw', y='ECG_Clean')
```
